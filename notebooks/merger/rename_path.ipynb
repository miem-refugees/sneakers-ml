```python
import pandas as pd
from pathlib import Path
```

```python
high = pd.read_csv("data/raw/metadata/highsnobiety.csv")
kick = pd.read_csv("data/raw/metadata/kickscrew.csv")


def fix_path(string):
    if pd.notnull(string):
        a = list(Path(string).parts)
        a.insert(2,"images")
        return str(Path(*a))
    else:
        return string

fix_path(high["right-side-img"][0])
```

```python
def rename_columns_df(df):
    columns = ["right-side-img","left-side-img","front-both-img"]
    for column in columns:
        df[column] = df[column].apply(fix_path)
    return df
```

```python
new_high = rename_columns_df(high)
new_high
```

```python
new_kick = rename_columns_df(kick)
new_kick
```

```python
new_high.to_csv("data/raw/metadata/new_highsnobiety.csv", index=False)
new_kick.to_csv("data/raw/metadata/new_kickscrew.csv", index=False)
```
